```python
# In Jupyter
from midi_handler import MIDIHandler

# Test connection
midi = MIDIHandler(knob_sensitivity=0.1)
midi.connect()

print("\nTurn knobs and press buttons...")
print("Press Ctrl+C to stop\n")

try:
    while True:
        event = midi.read_event()
        
        if event:
            if event.control_type == 'knob':
                print(f"Knob {event.control_id}: delta={event.delta:+3d}")
            elif event.control_type == 'button':
                print(f"Button {event.control_id} pressed")
                
except KeyboardInterrupt:
    print("\nStopping...")
    midi.disconnect()
```
