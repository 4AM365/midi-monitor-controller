```python
# In Jupyter
from monitor_controller import MonitorController
```

```python
# Test basic operations
with MonitorController() as monitor:
    # Get current brightness
    current = monitor.get_brightness()
    print(f"Current brightness: {current}")
    
    # Set brightness to 50
    print("Setting brightness to 50...")
    monitor.set_brightness(50)
    
    # Verify
    new = monitor.get_brightness()
    print(f"New brightness: {new}")
    
    # Restore original
    monitor.set_brightness(current)
    print(f"✓ Restored to {current}")
```
