```python
from virtual_knob import VirtualKnob
```

```python
# Test in Jupyter


# Create brightness knob
brightness = VirtualKnob(
    name="Brightness",
    initial=75,
    min_val=0,
    max_val=100
)

print(f"Initial: {brightness}")

# Simulate X-Touch sending inverted values
# Physical clockwise turn → value=1 → delta = -(1-64) = +63
# We want clockwise to DECREASE brightness

def xtouch_to_delta(cc_value: int) -> int:
    """Convert X-Touch CC value to delta (inverted)"""
    # Your X-Touch: clockwise = low values (1-10)
    # We want: clockwise = decrease brightness
    
    raw_delta = -(cc_value - 64)  # Invert the standard formula
    
    # Now we need to flip the sign for brightness
    # Clockwise (positive delta from formula) should subtract
    return -raw_delta

# Test clockwise turn (X-Touch sends value=1)
delta = xtouch_to_delta(1)  # Should be negative (decrease)
new_val, hit_limit = brightness.adjust(delta)
print(f"Clockwise (CC=1): delta={delta}, new={new_val}, limit={hit_limit}")

# Test counter-clockwise (X-Touch sends value=127)
delta = xtouch_to_delta(127)  # Should be positive (increase)
new_val, hit_limit = brightness.adjust(delta)
print(f"Counter-CW (CC=127): delta={delta}, new={new_val}, limit={hit_limit}")
```
